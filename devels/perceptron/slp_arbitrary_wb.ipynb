# perceptron classification ability

```python
def binary_step(x):
    if x >= 0:
        z = 1
    else:
        z = 0
    return z

def perceptron(w, x):
    ww = np.array(w)
    xx = np.array(x + [1])
    y = binary_step(np.dot(ww, xx))
    return y
```

```python
import numpy as np

weights = [1, 2, 1]

x = [5, 0]
y = perceptron(weights, x)
print(y)

print()

x = [5, -5]
y = perceptron(weights, x)
print(y)
```

## decision boundary line

```python
def decision_boundary(xr, yr, w):
    a = w[0]; b = w[1]; c = w[2]
    xmin = xr[0]; xmax = xr[1]
    ymin = yr[0]; ymax = xr[1]

    xx = []; yy = []
    for x in range(xmin, xmax+1):
        for y  in range(ymin, ymax+1):
            z = a*x + b*y + c
            if z == 0:
                xx.append(x)
                yy.append(y)
    return xx, yy
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipynb.fs.full.datasets2 as ds2

w = [1, -2, -1]
xx = []; yy = []; zz = []
for x2 in range(0, 11):
    for x1 in range(0, 21):
        x = [x1, x2]
        y = perceptron(w, x)
        xx.append(x1)
        yy.append(x2)
        zz.append(y)

xr = [0, 20]; yr = [0, 10]
xb, yb = decision_boundary(xr, yr, w)
ds2.plot_dataset(plt, xx, yy, zz)
plt.plot(xb, yb, 'g--', lw=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipynb.fs.full.datasets2 as ds2

w = [1, -2, -1]

xr = [0, 21]; yr = [0, 11]
xb, yb = decision_boundary(xr, yr, w)

plt.figure(figsize=(4, 4)); plt.grid()
plt.plot(xb, yb, 'g--', lw=2)
plt.xlim(0, 20); plt.ylim(0, 10)
plt.xlabel('x1'); plt.ylabel('x2');
plt.show()
```
