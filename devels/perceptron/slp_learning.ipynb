# perceptron learning process

## initial weights and bias

```python
import numpy as np
import matplotlib.pyplot as plt
import ipynb.fs.full.datasets2 as ds2
import ipynb.fs.full.model as nn

xr = [0, 20, 2]; yr = [0, 10, 1]; cs = [1, -1, 0]
xx, yy, zz = ds2.linear_separable(xr, yr, cs)
ds2.plot_dataset(plt, xx, yy, zz)

w = [2, 3, -30]
xb, yb = nn.decision_boundary(xr, yr, w)
plt.plot(xb, yb, 'g--', lw=2); plt.show()
```

## calculate difference between observed and predicted data

```python
import numpy as np
import ipynb.fs.full.model as nn

xr = [0, 20, 2]; yr = [0, 10, 1]; cs = [1, -1, 0]
xx, yy, zz = ds2.linear_separable(xr, yr, cs)

w = [2, 3, -30]; alpha = 0.5; epoch = 100
w1 = [w[0]]; w2 = [w[1]]; b = [w[2]]
for t in range(epoch):
    dw1 = 0; dw2 = 0; db = 0
    for x1, x2, y in zip(xx, yy, zz):
        x = [x1, x2]
        yp = nn.perceptron(w, x)
        w[0] = w[0] + alpha * (y - yp) * x1
        w[1] = w[1] + alpha * (y - yp) * x2
        w[2] = w[2] + alpha * (y - yp) * 1
    
    w1.append(w[0]); w2.append(w[1]); b.append(w[2])
```

```python
import matplotlib.pyplot as plt

plt.plot(w1)
plt.plot(w2)
plt.plot(b)
plt.legend(['w1', 'w2', 'b'])
plt.xlabel('epoch')
plt.ylabel('params')
plt.grid()
plt.show()
```

## other case

```python
import numpy as np
import matplotlib.pyplot as plt
import ipynb.fs.full.datasets2 as ds2
import ipynb.fs.full.model as nn

xr = [0, 20, 2]; yr = [0, 10, 1];
cs = [1, -2, 4]
xx, yy, zz = ds2.linear_separable(xr, yr, cs)
ds2.plot_dataset(plt, xx, yy, zz)

w = [4, 5, -20]
xb, yb = nn.decision_boundary(xr, yr, w)
plt.plot(xb, yb, 'g--', lw=2); plt.show()
```

```python
data = [xx, yy, zz]

w = [4, 5, -20]
alpha = 0.01; epoch = 500

w1, w2, b = nn.slp_learning(
    data, w, alpha, epoch)

plt.plot(w1)
plt.plot(w2)
plt.plot(b)
plt.legend(['w1', 'w2', 'b'])
plt.xlabel('epoch')
plt.ylabel('params')
plt.grid(); plt.show()
```

```python
ds2.plot_dataset(plt, xx, yy, zz)

wf = [w1[-1], w2[-1], b[-1]]
xb, yb = nn.decision_boundary(xr, yr, wf)
plt.plot(xb, yb, 'g--', lw=2); plt.show()
```
